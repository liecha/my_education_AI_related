```python
def split_dataset(datafram_indata):

    # Pick out day 12 of januari as test values
    day_2022_02_12 = datafram_indata[datafram_indata['Datum Ford'] == "2022-02-12"]
    day_2022_02_13 = datafram_indata[datafram_indata['Datum Ford'] == "2022-02-13"]
    frames = [day_2022_02_12, day_2022_02_13]
    test = pd.concat(frames)
    test = test['AntalBokadeFramkörning'].values

    # Select the set of data for training (2019-07-01 to 2022-01-21)
    train = datafram_indata[datafram_indata['Datum Ford'] != "2022-02-12"]
    train = train[train['Datum Ford'] != "2022-02-13"]
    train = train['AntalBokadeFramkörning'].values
    train = train.reshape(-1, 1)

    return train, test
```

```python
train, test = split_dataset(datafram_indata)
```

```python
def series_to_supervised():

    # Implement a scaler with value (range) between 0 and 1
    scaler = MinMaxScaler(feature_range = (0,1))

    # Apply scaler in training set
    training_set_scaled = scaler.fit_transform(train)

    # Create a data structure with 48 timesteps and 1 output
    # timesteps --> 288  previous timeslots are used to predict the next output
    # X_train --> 288 previous temperature measures
    # y_traing --> value at the current day
    # NOTE: [] --> type list NOT np array
    X_train = []
    y_train = []

    for i in range(timeslot, len(train)):
        X_train.append(training_set_scaled[i-timeslot:i,0])
        y_train.append(training_set_scaled[i,0])

    # Convert the lists X_train and y_train to np.arrays
    X_train, y_train = np.array(X_train), np.array(y_train)

    # Reshape data to fulfill requirement of inputs for the RNN 
    # RNN == (Recurrent Neural Network)
    # We currently have an 2D matrix - X_train --> (1196, 60) matrix
    # We want 3D matrix - the third dimension corresponds to the indicator
    # In this case the indicator is 1 since we have one output
    # array.shape[0] --> antal rader
    # array.shape[1] --> antal kolumner
    # reshape(arrayLike, newShape, order)
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

    return X_train, y_train, scaler
```

```python
X_train, y_train, scaler = series_to_supervised()
```

```python
def RNN(X_train, neurons):
    
    # Initializing the RNN 
    # The output is continous value --> use regression
    regressor = Sequential()
    
    # Add the first layer to the neural network
    regressor.add(LSTM(units = neurons, return_sequences = True, input_shape=(X_train.shape[1], X_train.shape[2])))
    
    # Dropout regulation (use to be 20%)
    # --> this is the number of neurons to be ignored
    regressor.add(Dropout(0.2))
    
    # Add more layers to the neural network
    # Second layer
    regressor.add(LSTM(units = neurons, return_sequences = True))
    regressor.add(Dropout(0.2))
    
    # Third layer
    regressor.add(LSTM(units = neurons, return_sequences = True))
    regressor.add(Dropout(0.2))
    
    # Forth layer - last layer before output  layer
    regressor.add(LSTM(units = neurons))
    regressor.add(Dropout(0.2))
    
    # Output layer
    regressor.add(Dense(units = 1))
    
    # Compiling the RNN
    # NOTE: Optimizer - RMSprops is recommended for RNN but Adam was detected to 
    # be a better choice for this problem
    regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')
    
    return regressor
```

```python
    def train_neural_network(epochs, batch_size, neurons):
        
        # Buildning the ANN (Artificial Neural Network)
        # Create a ANN model instance
        rnn = RNN(X_train, neurons)
               
        # Training - fit the ANN to the training set
        history = rnn.fit(
            X_train, 
            y_train, 
            epochs = epochs, 
            batch_size = batch_size)
        
        return rnn, history
```

```python
rnn, history = train_neural_network(epochs, batch_size, neurons)
```

```python
def rnn_prediction(datafram_indata, scaler, rnn):

    # Pick out day 12 of januari as test values
    day_2022_02_12 = datafram_indata[datafram_indata['Datum Ford'] == "2022-02-12"]
    day_2022_02_13 = datafram_indata[datafram_indata['Datum Ford'] == "2022-02-13"]
    frames = [day_2022_02_12, day_2022_02_13]
    test = pd.concat(frames)

    # Select the set of data for training (2019-07-01 to 2022-01-21)
    train = datafram_indata[datafram_indata['Datum Ford'] != "2022-02-12"]
    train = train[train['Datum Ford'] != "2022-02-13"]

    dataset_total = pd.concat((train['AntalBokadeFramkörning'], test['AntalBokadeFramkörning']), axis = 0)
    inputs = dataset_total[len(dataset_total) - len(test) - timeslot:].values
    inputs = inputs.reshape(-1,1)
    inputs = scaler.transform(inputs)

    X_test = []
    for i in range(timeslot, timeslot * 2):
        X_test.append(inputs[i-timeslot:i, 0])
    X_test = np.array(X_test)
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

    # Predicting the Test set results
    # NOTE: be careful to choose the right regressor here!
    # I have tried out the load function and I predict the air temp from
    # the saved RNN here (new_model instead of regressor)
    predicted_values = rnn.predict(X_test)
    prediction_unscaled = scaler.inverse_transform(predicted_values)

    return prediction_unscaled
```

```python
prediction_unscaled = rnn_prediction(datafram_indata, scaler, rnn)
```

```python
def plot_result(datafram_indata):

    # Pick out day 22 and 23 of januari as test values
    day_2022_02_12 = datafram_indata[datafram_indata['Datum Ford'] == "2022-02-12"]

    date_time = pd.to_datetime(day_2022_02_12['Datum Ford'] + ' ' + day_2022_02_12['Tid Ford'])

    # Prepare data for the plots
    # We want to show predicted stops as 1 and actual stops as -1
    # We want all other valus to be nan --> not shown in the plot
    # This is because we are only interested in stops - not movment
    y_pred_booking = []
    for i in range(0, len(prediction_unscaled)):
        if prediction_unscaled[i] > 0.4:
            y_pred_booking.append(-1)
        else:
            y_pred_booking.append(0)

    y = day_2022_02_12['AntalBokadeFramkörning'].values
    y_test_booking = []

    for i in range(0, len(y)):
        if y[i] == 0:
            y_test_booking.append(0)
        else:
            y_test_booking.append(1)


    # Visualising the results      
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    fig.autofmt_xdate(rotation=45)
    ax.plot(date_time[0:228], y_test_booking[0:228])
    ax.plot(date_time[0:228], y_pred_booking[0:228])
    xftm = mdates.DateFormatter('%y-%m-%d %H:%M')
    ax.xaxis.set_major_formatter(xftm)

    ax.grid(True)

    plt.title('Prediktion Antal Bokningar och Framkörningar')
    plt.xlabel('Time')
    plt.ylabel('Antal Bokningar/Framkörningar')
    plt.legend(['Faktiskt bokning/framkörning', 'Predikterad bokning/framkörning'])
```

```python
plot_result(datafram_indata)
```
